Mostrar los tipos de accidente por cada estado


```python
import pandas as pd

# Definir la ruta del archivo CSV
file_path = "/kaggle/input/accidentes1/conjunto_de_datos/atus_anual_2022.csv"

# Cargar los datos desde el archivo CSV
data = pd.read_csv(file_path)

# Cargar la lista de ID de entidad y nombres de entidad
entidades = {
    1: 'Aguascalientes', 2: 'Baja California', 3: 'Baja California Sur', 4: 'Campeche',
    5: 'Coahuila de Zaragoza', 6: 'Colima', 7: 'Chiapas', 8: 'Chihuahua', 9: 'Ciudad de México',
    10: 'Durango', 11: 'Guanajuato', 12: 'Guerrero', 13: 'Hidalgo', 14: 'Jalisco', 15: 'México',
    16: 'Michoacán de Ocampo', 17: 'Morelos', 18: 'Nayarit', 19: 'Nuevo León', 20: 'Oaxaca',
    21: 'Puebla', 22: 'Querétaro', 23: 'Quintana Roo', 24: 'San Luis Potosí', 25: 'Sinaloa',
    26: 'Sonora', 27: 'Tabasco', 28: 'Tamaulipas', 29: 'Tlaxcala', 30: 'Veracruz de Ignacio de la Llave',
    31: 'Yucatán', 32: 'Zacatecas'
}

# Agrupar por entidad y contar los tipos de accidente
tipos_por_estado = data.groupby(['ID_ENTIDAD', 'TIPACCID']).size().reset_index(name='count')

# Cambiar el ID de entidad por el nombre de la entidad
tipos_por_estado['ID_ENTIDAD'] = tipos_por_estado['ID_ENTIDAD'].map(entidades)

# Pivotar los datos para obtener una tabla donde las columnas son los tipos de accidente
tabla_tipos_por_estado = tipos_por_estado.pivot(index='ID_ENTIDAD', columns='TIPACCID', values='count')

# Reemplazar NaN por 0
tabla_tipos_por_estado = tabla_tipos_por_estado.fillna(0)

# Mostrar la tabla
print(tabla_tipos_por_estado)
```

Mostrar la cantidad de accidentes por dia de la semana en sinaloa por edad del conductor

```python
import pandas as pd

# Definir la ruta del archivo CSV
file_path = "/kaggle/input/accidentes1/conjunto_de_datos/atus_anual_2022.csv"

# Cargar los datos desde el archivo CSV
data = pd.read_csv(file_path)

# Filtrar los datos para incluir solo accidentes en Sinaloa (ID_ENTIDAD = 25)
sinaloa_data = data[data['ID_ENTIDAD'] == 25]

# Agrupar por día de la semana, edad del conductor y tipo de accidente, y contar los accidentes
accidentes_por_dia_edad_tipo = sinaloa_data.groupby(['DIASEMANA', 'ID_EDAD', 'TIPACCID']).size().reset_index(name='count')

# Mostrar los resultados
print(accidentes_por_dia_edad_tipo)
```

Mostrar el ranking de accidentes por mes por estado en una animacion que muestre desde 1997 hasta el 2022

```python
import pandas as pd

# Definir la ruta base del archivo CSV
base_file_path = "/kaggle/input/accidentes1/conjunto_de_datos/atus_anual_"

# Lista para almacenar los DataFrames de cada año
dfs = []

# Iterar desde 1997 hasta 2022
for year in range(1997, 2023):
    # Construir la ruta completa del archivo CSV para el año actual
    file_path = f"{base_file_path}{year}.csv"
    
    # Cargar los datos desde el archivo CSV
    try:
        data = pd.read_csv(file_path)
    except FileNotFoundError:
        continue  # Si el archivo no existe para este año, pasar al siguiente
    
    # Calcular el ranking de accidentes por mes y estado para el año actual
    monthly_ranking = data.groupby(['MES', 'ID_ENTIDAD']).size().reset_index(name='accidents')
    monthly_ranking['rank'] = monthly_ranking.groupby('MES')['accidents'].rank(ascending=False, method='min')
    
    # Agregar el DataFrame del año actual a la lista
    dfs.append(monthly_ranking)

# Combinar todos los DataFrames en uno solo
combined_data = pd.concat(dfs, ignore_index=True)

# Mostrar el DataFrame combinado
print(combined_data)
```

Mostrar la cantidad de accidentes por municipio del estado de sinaloa en los ultimos 10 años

```python
import pandas as pd

# Definir la ruta base del archivo CSV
base_file_path = "/kaggle/input/accidentes1/conjunto_de_datos/atus_anual_"

# Lista para almacenar los DataFrames de cada año
dfs = []

# Iterar desde 2013 hasta 2022 (últimos 10 años)
for year in range(2013, 2023):
    # Construir la ruta completa del archivo CSV para el año actual
    file_path = f"{base_file_path}{year}.csv"
    
    # Cargar los datos desde el archivo CSV
    try:
        data = pd.read_csv(file_path)
    except FileNotFoundError:
        continue  # Si el archivo no existe para este año, pasar al siguiente
    
    # Filtrar los datos solo para el estado de Sinaloa
    sinaloa_data = data[data['ID_ENTIDAD'] == 25]  # ID_ENTIDAD 25 corresponde a Sinaloa
    
    # Agregar el DataFrame del año actual a la lista
    dfs.append(sinaloa_data)

# Combinar todos los DataFrames en uno solo
combined_data = pd.concat(dfs, ignore_index=True)

# Calcular la cantidad de accidentes por municipio
accidents_by_municipality = combined_data.groupby('ID_MUNICIPIO').size().reset_index(name='accidents')

# Mostrar la cantidad de accidentes por municipio en Sinaloa en los últimos 10 años
print(accidents_by_municipality)
```

Mostrar la cantidad de accidentes distinguidos entre dia festivo, fin de semana y dia normal

```python
import pandas as pd

# Cargar los datos de accidentes
accidentes = pd.read_csv("/kaggle/input/accidentes1/conjunto_de_datos/atus_anual_2022.csv")

# Convertir las columnas ANIO, MES e ID_DIA a tipo str y concatenarlas para formar la fecha
accidentes['FECHA'] = pd.to_datetime(accidentes['ANIO'].astype(str) + '-' + accidentes['MES'].astype(str) + '-' + accidentes['ID_DIA'].astype(str), errors='coerce')

# Filtrar las filas con fechas válidas
accidentes_validos = accidentes.dropna(subset=['FECHA'])

# Definir una lista con las fechas de los días festivos
dias_festivos = ['2022-01-01', '2022-02-05', '2022-03-21', '2022-05-01', '2022-09-16', '2022-12-25']

# Definir una función para determinar el tipo de día
def tipo_dia(fecha):
    if fecha.weekday() >= 5:
        return 'Fin de semana'
    elif fecha.strftime('%Y-%m-%d') in dias_festivos:
        return 'Día Festivo'
    else:
        return 'Día Normal'

# Agregar una columna para el tipo de día
accidentes_validos['Tipo de Día'] = accidentes_validos['FECHA'].apply(tipo_dia)

# Contar la cantidad de accidentes por tipo de día
accidentes_por_tipo_dia = accidentes_validos['Tipo de Día'].value_counts().reset_index()

print(accidentes_por_tipo_dia)
```
