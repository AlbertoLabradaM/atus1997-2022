Se crea un dataframe con los datos de los accidentes de los ultimos 3 años

```python
df = pd.DataFrame()
years=[2020,2021,2022]

#Se crean un dataframe por cada uno de los años para poder observar su contenido uno a uno
df_2020=pd.read_csv("../input/atus_anual_2020.csv")
df_2021=pd.read_csv("../input/atus_anual_2021.csv")
df_2022=pd.read_csv("../input/atus_anual_2022.csv")

#Se crea un dataframe con todos los años dentro de un for
for year in years:
    filename = f"../input/atus_anual_{year}.csv"
    data = pd.read_csv(filename)  
    df = pd.concat([df,data], ignore_index=False)  


df.head()

```

Se agrega una columna tipo date que englobe las columnas correspondientes a Año, Mes, Dia, Hora y Minuto.
Debido a que en la columna de dia hay una incompatibilidad de tipo de dato en el dataset del 2020 y 2021 con respecto al de  2022, solo es posible realizar el agregado de la columna fecha en este ultimo

```python
df_2022['FECHA'] = pd.to_datetime(df_2022[['ANIO', 'MES', 'ID_DIA', 'ID_HORA', 'ID_MINUTO']].astype(str).agg('-'.join, axis=1), format='%Y-%m-%d-%H-%M')
df_2022.head()
```

 Se agrega una columna para Área cuyos valores serán Urbana o Suburbana dependiendo del área donde ocurrió el accidente.

```python
df['ÁREA'] = 'Urbana'
df.loc[df['URBANA'] == 'Sin accidente en esta zona', 'ÁREA'] = 'Suburbana'
df
```

Se elimina las columnas URBANA y SUBURBANA.
En el dataframe que contiene los 3 datasets no existe la columna SUBURBANA, por lo que la instruccion solo elimina la columna de urbana

```python
df=df.drop(columns=['URBANA','SUBURBANA'],axis=1)
df
```

Sin embargo en el dataframe de los datos del 2022 si contiene, por lo que se ejecuta y elimina dichas columnas

```python
df_2022=df_2022.drop(columns=['URBANA','SUBURBANA'],axis=1)
df_2022
```

Se determina cual municipio de cada estado tuvo más meses sin accidentes

```python
#Crea una columna que combina la informacion de las columnas año y mes
df_2022['FECHA_MES'] = pd.to_datetime(df_2022[['ANIO', 'MES']].astype(str).agg('-'.join, axis=1))

# Agrupa los datos en el DataFrame df_2022 por 'ID_ENTIDAD' y 'ID_MUNICIPIO', y luego 
#cuenta el número único de meses sin accidentes ('FECHA_MES') para cada par de entidad y municipio.
sin_accidentes = df_2022.groupby(['ID_ENTIDAD', 'ID_MUNICIPIO'])['FECHA_MES'].nunique().reset_index()

# Esta línea de código encuentra el municipio con la menor cantidad de meses sin accidentes para cada estado,
# seleccionando el índice correspondiente utilizando el método idxmin() dentro de cada grupo de 'ID_ENTIDAD'.
# Luego, se utiliza loc[] para seleccionar las filas correspondientes a estos índices.
mun_menos_accidentes = sin_accidentes.loc[sin_accidentes.groupby('ID_ENTIDAD')['FECHA_MES'].idxmin()]

#Imprime el mensaje y muestra la tabla
print("Muninicipio con menos accidentes por estado.")
mun_menos_accidentes
```

Se resume los accidentes del último año por tipo y que porcentaje representan

```python
#Se guarda el año que hay dentro de df
last_year= df['ANIO'].max()

#Se guardan los accidentes que hay de ese ultimo año
accidentes_LY = df[df['ANIO'] == last_year]

# Agrupa los registros filtrados por el tipo de accidente ('TIPACCID') y cuenta la cantidad de accidentes de cada tipo.
accidentes = accidentes_LY.groupby('TIPACCID').size().reset_index(name='COUNT')

# Calcula el total de accidentes sumando los valores de la columna 'COUNT' en el DataFrame 'accidentes'.
total = accidentes['COUNT'].sum()
#Se calcula porcentaje tomando en cuenta el total
accidentes['PORCENTAJE'] = (accidentes['COUNT'] / total) * 100

accidentes
```

Se obtiene la cantidad total de muertos y lesionados en los ultimos 3 años


```python
#Se obtiene la suma de todos los tipos de muertos de cada año y se guardan en una variable
total_muertos_2020=(df_2020['CICLMUERTO']!=0).sum()+(df_2020['CONDMUERTO'] !=0).sum()+(df_2020['NEMUERTO'] !=0).sum()+(df_2020['PEATMUERTO'] !=0).sum()+(df_2020['PASAMUERTO']!=0).sum()+(df_2020['OTROMUERTO']!=0).sum()
total_muertos_2021=(df_2021['CICLMUERTO']!=0).sum()+(df_2021['CONDMUERTO'] !=0).sum()+(df_2021['NEMUERTO'] !=0).sum()+(df_2021['PEATMUERTO'] !=0).sum()+(df_2021['PASAMUERTO']!=0).sum()+(df_2021['OTROMUERTO']!=0).sum()
total_muertos_2022=(df_2022['CICLMUERTO']!=0).sum()+(df_2022['CONDMUERTO'] !=0).sum()+(df_2022['NEMUERTO'] !=0).sum()+(df_2022['PEATMUERTO'] !=0).sum()+(df_2022['PASAMUERTO']!=0).sum()+(df_2022['OTROMUERTO']!=0).sum()
total_muertos=total_muertos_2020+total_muertos_2021+total_muertos_2022

#Se obtiene la suma de todos los tipos de heridos de cada año y se guardan en una variable
total_heridos_2020=(df_2020['NEHERIDO']!="Sólo daños").sum()+(df_2020['CONDHERIDO']!=0).sum()+(df_2020['PASAHERIDO']!=0).sum()+(df_2020['PEATHERIDO']!=0).sum()+(df_2020['OTROHERIDO']).sum()
total_heridos_2021=(df_2021['NEHERIDO']!="Sólo daños").sum()+(df_2021['CONDHERIDO']!=0).sum()+(df_2021['PASAHERIDO']!=0).sum()+(df_2021['PEATHERIDO']!=0).sum()+(df_2021['OTROHERIDO']).sum()
total_heridos_2022=df_2022['NEHERIDO'].sum()+(df_2022['CONDHERIDO']!=0).sum()+(df_2022['PASAHERIDO']!=0).sum()+(df_2022['PEATHERIDO']!=0).sum()+(df_2022['OTROHERIDO']).sum()
total_heridos=total_heridos_2020+total_heridos_2021+total_heridos_2022

print("Muertos en los ultimos 3 años: ", total_muertos,"\n")
print("Heridos en los ultimos 3 años: ", total_heridos)
```

¿Cómo se compara el número de accidentes de cada año contra el anterior? ¿En qué porcentaje varió?

```python
#Se obtiene la cantidad de accidentes por año (cantidad de filas registradas por año)
accidentes_2020=df_2020.shape[0]
accidentes_2021=df_2021.shape[0] 
accidentes_2022=df_2022.shape[0]

#Se obtiene la diferencia entre un año y otro, así como el porcentaje de variación que se tiene
diferencia_20_21=accidentes_2021 - accidentes_2020
variacion_20_21=diferencia_20_21 / accidentes_2020 * 100

diferencia_21_22=accidentes_2022 - accidentes_2021
variacion_21_22=diferencia_21_22 / accidentes_2021 * 100

#Se muestra el total de accidentes por año
print("Total de accidentes en 2020: ",accidentes_2020)
print("Total de accidentes en 2021: ",accidentes_2021)
print("Total de accidentes en 2022: ",accidentes_2022)

#Se muestra la diferencia entre un año y otro, así como el respectivo porcentaje de variacion que tiene
print("Existe una diferencia de ",(diferencia_20_21)," accidentes entre el 2020 y 2021")
print("Existe un ",round(variacion_20_21,2),"% de variacion entre los accidentes ocurridos entre el 2020 y 2021")
print("Mientras que")
print("Existe una diferencia de ",(diferencia_21_22)," accidentes entre el 2021 y 2022")
print("Existe un ",round(variacion_21_22,2),"% de variacion entre los accidentes ocurridos entre el 2021 y 2022")
```
